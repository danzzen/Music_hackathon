```python
import numpy as np # linear algebra
import pandas as pd
```

```python
 # data processing, CSV file I/O (e.g. pd.read_csv)
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re
import string
# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directo
data=pd.read_csv('tripadvisor_co_uk-travel_restaurant_reviews_sample.csv')#read data
data.head()#print dataframe head
data.drop(data[data['review_text']=='NaN'].index,inplace=True)#remove coloums that have nulll values
stops = set(stopwords.words("english"))#inilize stopword
```

```python
##create a function to clean data it includes stopword removal,tokenizing and part of speech tagging
#
def cleanData(text, lowercase = False, remove_stops = False,pos= False):
    txt = str(text)
    txt = re.sub(r'[^A-Za-z0-9\s]',r'',txt)
    txt = re.sub(r'\n',r' ',txt)

    if lowercase:
        txt = " ".join([w.lower() for w in txt.split()])

    if remove_stops:
        txt = " ".join([w for w in txt.split() if w not in stops])

    if pos:
        txt = ''.join(txt)
        txt = word_tokenize(txt)
        word_tag_pair = nltk.pos_tag(txt)

    return word_tag_pair
```

```python
data['review_text'] = data['review_text'].map(lambda x: cleanData(x, lowercase=False, remove_stops=True,pos = True))##using map function cleandata function is applied  on every data['review_text] element
```

```python
import ast
```

```python
def aspect_extraction(word_tag_pair,output):
    inputTupples=ast.literal_eval(str(word_tag_pair))
    prevWord=''
    prevTag=''
    currWord=''
    aspectList=[]
    outputDict={}
    # print(inputTupples)
    #Extracting Aspects
    # print(inputTupples[0][0])
    for value in inputTupples:
            if(value[1]=='NN' or value[1]=='NNP'):
                if(prevTag=='NN' or prevTag=='NNP'):
                    currWord= prevWord + ' ' + value[0]
                else:
                    aspectList.append(prevWord.upper())
                    currWord= value[0]
            prevWord=currWord
            prevTag=value[1]
    #Eliminating aspect which has 1 or less count

    for aspect in aspectList:
            if(aspectList.count(aspect)>1):
                    if(outputDict.keys()!=aspect):
                            outputDict[aspect]=aspectList.count(aspect)
    outputAspect=sorted(outputDict.items(), key=lambda x: x[1],reverse = True)
    if(len(outputAspect)>1):
        output.append(aspectList)
```

```python
output = []
for i in range(0,len(data['review_text'])):
    aspect_extraction(data['review_text'].iloc[i],output)
```

```python
final_aspect_list = np.array(output)
```
